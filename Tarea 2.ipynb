# *Siniestralidad de la cuenta GZ*

El archivo importando tiene recopilacin de la  informacin de siniestralidad de la cuenta GZ, en el cual, nos muestra la causa de los accidentes, la cobertura afectada de la pliza de autos y el monto pagado.  

```python
df_csv = pd.read_csv(r"C:\Users\irwin\Downloads\Sin_GZ.csv")
df_csv
```

# Primeras filas

```python
df_csv.head()
```

```python
df_csv['COSTO NETO TOTAL'].head(10)
```

```python
# La columna COSTO NETO TOTAL en formato string

# 1. Remplaza la coma separadora de miles por nada (la elimina)
df_csv['COSTO NETO TOTAL'] = df_csv['COSTO NETO TOTAL'].str.replace(',', '')

# 2. Convierte la columna a numrico (float porque tiene decimales)
df_csv['COSTO NETO TOTAL'] = pd.to_numeric(df_csv['COSTO NETO TOTAL'], errors='coerce')
```

# Informacin general de las variables

```python
df_csv.dtypes
```

# Eliminar las columnas CDR, RED y TIPO TALLER/AGENCIA

```python
df_csv = df_csv.drop(columns=['CDR'], errors='ignore')
df_csv.head()

df_csv = df_csv.drop(columns=['RED'], errors='ignore')
df_csv.head()

df_csv = df_csv.drop(columns=['TIPO TALLER/AGENCIA'], errors='ignore')
df_csv.head()

```

# Filtar siniestros mayores a 30 mil pesos 

```python
filtro_SN = df_csv[df_csv['COSTO NETO TOTAL'] > 30000]

print('\nSiniestros con COSTO NETO TOTAL mayor a  30 mil pesos:')
display(filtro_SN)
```

# Subconjuntos de datos
## EL asegurado No es responsable en el siniestro

```python
No_responsable= df_csv[df_csv['RESPONSABILIDAD'] == 'NO']
print('\nSubconjunto - No Responsable del siniestro:')
display(No_responsable)
     
```

# Agrupar por la columna 'RESPONSABILIDAD' y contar cuntos siniestros hay en cada grupo



```python
grupo = df_csv.groupby('RESPONSABILIDAD').size()

print(grupo)
```
