# **Prueba de Normalidad Shapiro-Wilk**

## Realizamos hipotesis para determinar si el conjunto de datos que se distribuye de manera nomarl 

Hipótesis nula (H₀): Provienen de una distribución normal.

Hipótesis alternativa (Ha): No provienen de una distribución normal.

```python
# Seleccionamos solo las variables numéricas
variables_numericas = [
    "RESERVA",
    "AJUSTES",
    "PAGO",
    "COSTO NETO TOTAL",
    "CNS_RC"]
```

```python
# Prueba de normalidad Shapiro-Wilk
resultados = {}

for var in variables_numericas:
    datos =df_csv[var].dropna()  # eliminar NaN
    stat, p = shapiro(datos)
    resultados[var] = p

print(resultados)

normalidad = pd.DataFrame(resultados.items(), columns=["Variable", "p_value"])

# Clasificación según p-value
normalidad["Distribución"] = normalidad["p_value"].apply(lambda p: "Paramétrica" if p > 0.05 else "No paramétrica ")

display(normalidad)
```

# **Conclusiones de la matriz de correlación**

La correlación mide la relación lineal entre dos variables numéricas.
* Va de -1 a 1
* Cercano a 1: relación positiva fuerte (ambas suben o bajan juntas)
* Cercano a -1: relación negativa fuerte (una sube y la otra baja)
* Cercano a 0: poca o ninguna relación lineal

Por lo tanto: 

1. El COSTO NETO TOTAL depende fuertemente de los AJUSTES.
Cuando pasa un siniestro el ajustador da un monto aproximado que se debe de pagar este puede ser diferete casi igual o igual al Costo Neto Total.
2. CNS_RC está casi perfectamente correlacionado con el costo neto total.
Es la columna el cual es el monto final que se paga al afectado, solamente de una cobertura de la póliza.
3. RESERVA muestra una relación débil con las demás variables.
La Reserva es un cálculo en el cual se tiene un monto congelado, es decir, una estimación de lo que puede llegar a pagarse en un siniestro medio.
